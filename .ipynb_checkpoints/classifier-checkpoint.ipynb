```python
data = torch.load("training_data_tensors.pt")
train_ds = TensorDataset(data['x'], data['y'])
train_loader = DataLoader(train_ds, batch_size = 50, shuffle=True)
print(len(data['x']))
```

```python
# torch MLP setup

mlp = MLP_NN()


optimizer = optim.Adam(mlp.parameters(), lr= 0.0001)
criterion = nn.BCELoss()
# criterion = nn.CrossEntropyLoss()
```

```python

num_epochs = 200
mlp.train()
loss_over_time = []

for epoch in range(num_epochs): 
    running_loss = 0.0

    for data, target in train_loader:
        data = data.flatten(1)
        
        optimizer.zero_grad()
        output = mlp(data)
        
        loss = criterion(output, target)
        loss.backward()
        optimizer.step()
        
        running_loss += loss.item()

    print(f"Training loss {epoch}: {running_loss}")
    loss_over_time.append(running_loss)

print('Finished Training')

torch.save(mlp.state_dict(), 'saved_mlp_data.pth') 

'''
torch.save({
    "model_state": mlp.state_dict(),
    "optimizer_state": optimizer.state_dict(),
    "epoch": num_epochs,
    "loss_history": loss_history,
}, "saved_mlp_data.pth")
'''
```

```python
test_data = torch.load("test_data_tensors.pt")
test_ds = TensorDataset(test_data['x'], test_data['y'])
test_loader = DataLoader(test_ds, shuffle=True)
print(len(test_data['x']))


mlp.eval()

correct_predictions = 0
incorrect_class1 = 0
incorrect_class0 = 0

with torch.no_grad():          
    for data, label in test_loader:
        data = data.flatten(1)

        output = mlp(data)
        
        prediction = output
        
        if (prediction > 0.5 and label == 1) or (prediction < 0.5 and label == 0):
            correct_predictions += 1 
            
        else:
            if label == 1:
                incorrect_class1 += 1
            else:
                incorrect_class0 += 1



print('Model Accuracy: ', correct_predictions/len(test_data['x']))
print(incorrect_class1, ' wrong for natural languge', incorrect_class0, 'for noise')

```
