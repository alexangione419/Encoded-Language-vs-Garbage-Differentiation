```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import torch
import torchvision
import torch.nn as nn
import torch.optim as optim


import data_prep.obfuscator as obfuscator
import nl_processing.textEvaluator as textEvaluator
from mlp import MLP_NN
```

```python
# data preperation
collectedData = []
collectedLabels = []




lot = obfuscator.prepareAndObfuscateText("./data/natural_language/english_100k.txt")
for l in lot:
    collectedData.append(textEvaluator.textEvaluator(l))
    collectedLabels.append([1.0])

lot3 = obfuscator.prepareAndObfuscateText("./data/natural_language/german_100K.txt")
for l in lot3:
    collectedData.append(textEvaluator.textEvaluator(l))
    collectedLabels.append([1.0])


lot2 = obfuscator.prepareAndObfuscateText("./data/garbage.txt")
for l in lot2:
    collectedData.append(textEvaluator.textEvaluator(l))
    collectedLabels.append([0.0])


lot4 = obfuscator.prepareAndObfuscateText("./data/eng_news_2024_10K-sentences.txt")
for l in lot4:
    collectedData.append(textEvaluator.textEvaluator(l))
    collectedLabels.append([1.0])

lot5 = obfuscator.prepareAndObfuscateText("./data/ara-eg_web_2015_10K-sentences.txt")
for l in lot5:
    collectedData.append(textEvaluator.textEvaluator(l))
    collectedLabels.append([1.0])     

lot6 = obfuscator.prepareAndObfuscateText("./data/binary_segments.txt")
for l in lot6:
    collectedData.append(textEvaluator.textEvaluator(l))
    collectedLabels.append([0.0])   


lot7 = obfuscator.prepareAndObfuscateText("./data/eng-ca_web_2002_100K-sentences.txt")
for l in lot7:
    collectedData.append(textEvaluator.textEvaluator(l))
    collectedLabels.append([1.0])   

lot8 = obfuscator.prepareAndObfuscateText("./data/deu-ch_web_2002_100K-sentences.txt")
for l in lot8:
    collectedData.append(textEvaluator.textEvaluator(l))
    collectedLabels.append([1.0])  

lot9 = obfuscator.prepareAndObfuscateText("./data/gibberish.txt")
for l in lot9:
    collectedData.append(textEvaluator.textEvaluator(l))
    collectedLabels.append([0.0])  



print(len(collectedData))
print(len(collectedLabels))
```

```python
x_train = torch.tensor(collectedData, dtype=torch.float32)
y_train = torch.tensor(collectedLabels, dtype=torch.float32)

```

```python
# torch MLP setup

mlp = MLP_NN()


optimizer_ffn = optim.SGD(mlp.parameters(), lr= 0.1)
criterion = nn.BCELoss()
```

```python

num_epochs_ffn = 100

loss_over_time_nn = []
for epoch in range(num_epochs_ffn):  # loop over the dataset multiple times
    running_loss_ffn = 0.0

    for i, data in enumerate(x_train):
        # get the inputs; data is a list of [inputs, labels]
        
        # data = torch.flatten(data, start_dim=1)

        optimizer_ffn.zero_grad()

        output = mlp(data)

        loss = criterion(output, y_train[i])
        loss.backward()
        optimizer_ffn.step()
        running_loss_ffn += loss.item()

    print(f"Training loss {epoch}: {running_loss_ffn}")
    loss_over_time_nn.append(running_loss_ffn)

print('Finished Training')

torch.save(mlp.state_dict(), 'saved_mlp_data.pth')  # Saves model file (upload with submission)
```

```python
plt.plot(loss_over_time_nn, label="FFN Loss over time")
plt.show()
```

```python

mlp.eval()
lot = obfuscator.prepareAndObfuscateText("./data/french copy.txt")
data = textEvaluator.textEvaluator(lot[1])
print(data)

x_test = torch.tensor(data, dtype=torch.float32)


lot2 = obfuscator.prepareAndObfuscateText("./data/more_garbage.txt")
data2 = textEvaluator.textEvaluator(lot2[2])
print(data2)

x_test2 = torch.tensor(data2, dtype=torch.float32)


# feed forward nn calculation
outputs = mlp(x_test)
outputs2 = mlp(x_test2)
print(outputs2)
print(outputs)

if outputs.item() > 0.5:
    print("yay!")
```
